```python
# inference - predicting the house price from the size of the house

tf_price_pred = tf.add(tf.multiply(tf_size_factor, tf_house_size), tf_price_offset)
```

```python
# define loss function mean square

tf_error = tf.reduce_sum(tf.pow((tf_price_pred - tf_house_price),2)/(2*train_num))
```
